## American Airlines - Mobile App Reviews Data Collection
In this project, I will be analyzing customer feedback from reviews left on the American Airlines app from the Apple app store. First, we need to install the app store scraper in order to utilize the made API for web scraping. 

I'll include arguments to scrap data from the year 2020. In addition, we'll include a sleep argument so the scraping doesn't clutter web traffic of the website.

```python
import sys
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
from app_store_scraper import AppStore
import json
import datetime as dt
from pprint import pprint

import warnings
warnings.filterwarnings('ignore')
```

```python
aa_app = AppStore(
  country='us',       
  app_name='americanairlines',
  app_id=382698565    
) 
    

aa_app.review(
  how_many=15000, 
  after=dt.datetime(2017, 1, 1),
  sleep=np.random.randint(3,5)
)
```

```python
pprint(aa_app.reviews)
```

Now, we convert this raw data we managed to obtain into a pandas dataframe.

```python
aa_app.reviews
```

```python
aa_df = pd.DataFrame(columns=['username', 'review', 'isEdited', 'rating', 'title', 'date']) #username, review, isEdited, rating, title, date

userNames = []
reviews = []
isEditeds = []
ratings = []
titles = []
dates = []

for item in aa_app.reviews:
    user, rev, isEdit, rate, titl, dat = item['userName'], item['review'], item['isEdited'], item['rating'], item['title'], item['date']
    result = pd.Series({'username':user, 'review':rev, 'isEdited':isEdit, 'rating':rate, 'title':titl, 'date':dat}, name='x')
    
    aa_df = aa_df.append(result, ignore_index=False)
```

```python
aa_df_res = aa_df.reset_index().drop(columns='index')
```
